# Outliers

```python
print("Shape of USvideos data before removing outliers:",USvideos.shape)
Q1=USvideos.quantile(0.25)
Q3=USvideos.quantile(0.75)
IQR=Q3-Q1
print(IQR)
```

```python
USvideos_OutRemove = USvideos [ ~((USvideos< (Q1-1.5*IQR)  )| (  USvideos >   (Q3+1.5*IQR))).any(axis=1)   ];
print("Shape of USvideos after remove the outliers",USvideos_OutRemove.shape);
```
